```python
def signal_sample(t):
    return sin(1*pi*t) + sin(4*pi*t + 0.5)

N = 1000
f_s = 10
t = linspace(0,N/f_s,N)
f_t = signal_sample(t)     #
plt.plot(t,f_t)
plt.xlim(0,5)
plt.show()
```
