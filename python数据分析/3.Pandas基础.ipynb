## 4.maping and replace

```python
#北京：炸酱面，东京：寿司，首尔：拌饭
#增加新的一列
```

```python
df1_ = df1['首都'].map({'北京':'炸酱面','东京':'寿司','首尔':'炒肝'})
```

```python
df1['小吃'] = df1_
df1
```

```python
df1['小吃'].replace(['炸酱面','拌饭'],['炒肝','辣白菜'])
df1
```
